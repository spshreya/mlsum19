```
import csv
import matplotlib.pyplot as plt 
import pandas as pd
```

```
dff=pd.read_csv("bank.csv")
dff
```

```
creditscore=dff.loc[:, 'CreditScore']
age=dff.loc[:,'Age']
plt.bar(creditscore,age)
```

```
estsal=dff.loc[:, 'EstimatedSalary']
age=dff.loc[:,'Age']
'''plt.pie(Gender, labels = geography, radius = 1)
plt.legend()
plt.show()'''

plt.scatter(x=estsal,y=age )
```

```
f=0
m=0

for i in dff['Gender']:
    if i=='Male':
        m=m+1
    else:
        f=f+1
    

Gender=['Male','Female']
numgen=[m,f]
plt.pie( numgen,labels=Gender,autopct='%1.1f%%')
```

```
f=0
s=0
g=0

for i in dff['Geography']:
    if i=='France':
        f+=1
    elif i=='Spain':
        s+=1
    else:
        g+=1
    
#print(m,f)

Geo=['France','Spain','Germany']
place=[f,s,g]
exp=[0,0.1,0]
plt.pie(place,labels=Geo,explode=exp,autopct='%1.1f')
```
